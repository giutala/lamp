# KPI Calculation

Hereâ€™s the step-by-step explanation of the provided code:

### 1. **Import Libraries**
   - **pandas**: Used for data manipulation and analysis.
   - **numpy**: Provides support for large multi-dimensional arrays and matrices, and mathematical functions.
   - **matplotlib.pyplot**: A plotting library used for creating static, animated, and interactive visualizations.
   - **scipy.stats**: Contains statistical functions, such as the interquartile range (IQR), which is used for outlier detection.

### 2. **Data Preprocessing**
   - The **df.columns** line ensures that any new line characters (`\n`) in the column names are replaced with spaces. This step is important for cleaning the column names for easier processing.

### 3. **Initializing the Results Dictionary**
   - An empty list `monthly_kpis` is initialized to store Key Performance Indicators (KPIs) for each month and `equipaggio` (team).

### 4. **Iterating Over Groups of Data**
   - The **df.groupby(['Mese', 'Id Equipaggio'])** splits the DataFrame into groups based on the 'Mese' (month) and 'Id Equipaggio' (team).
   - For each group (`df_group`), the code checks if there are valid delta values for further processing.
   - If the group contains no valid data (i.e., `delta_values.empty`), it is skipped using `continue`.

### 5. **Step 1: Handling Outliers in Delta**
   - **IQR Calculation**: 
     - The first and third quartiles (`q1` and `q3`) are computed using `np.percentile`.
     - The **Interquartile Range (IQR)** is calculated using the `iqr()` function from `scipy.stats`.
     - **Outlier Boundaries**: The lower and upper bounds for valid data are computed as:
       - Lower bound = `q1 - 1.5 * IQR`
       - Upper bound = `q3 + 1.5 * IQR`
   - **Winsorizing**: The `filtered_delta` applies the `clip` function to limit values outside the IQR range to the respective bounds (this step mitigates the influence of extreme outliers).

### 6. **Step 2: Calculate Absolute Median Delta**
   - The **median** of the filtered delta values is calculated using `filtered_delta.median()`.
   - The result is then converted to an absolute value using `abs()`. This step normalizes the data to prevent the sign of the delta value from affecting the analysis.

### 7. **Step 3: Calculate Rescheduling Impact**
   - **Total Reschedules**: The reschedules are calculated by summing up the deliveries rescheduled for three different types of issues:
     - Type 1: Issues with rooms, wardrobes, and kids' rooms.
     - Type 2: Issues with kitchens and living rooms.
     - Type 3: Issues with bathroom furniture, tables, and chairs.
   - **Reschedule Ratio**: The ratio of reschedules to the total number of customers served is computed as:
     - \( \text{Reschedule Ratio} = \frac{\text{Total Reschedules}}{\text{Total Reschedules} + \text{Total Served}} \)

### 8. **Storing Results**
   - The results for each combination of month (`Mese`) and team (`Id Equipaggio`) are stored in the `monthly_kpis` list as a dictionary containing the 'Mese', 'Id Equipaggio', 'Delta Median', and 'Reschedule Ratio'.

### 9. **Convert to DataFrame**
   - The `monthly_kpis` list is converted into a DataFrame (`kpi_df`) and sorted by month and team.

### 10. **Normalization Parameters**
   - The minimum and maximum values for both the 'Delta Median' and 'Reschedule Ratio' columns are calculated. These values are used to normalize the data into a standardized range (0-100) for easier comparison.

### 11. **Weight Definition**
   - Weights (`W1` and `W2`) are defined for the two metrics. These weights reflect the relative importance of the 'Delta Median' and the 'Reschedule Ratio' in the final composite score.

### 12. **Normalization of Metrics**
   - The **Delta Median** and **Reschedule Ratio** are normalized using the formula:
     - \( \text{Normalized Value} = \frac{\text{Max Value} - \text{Current Value}}{\text{Max Value} - \text{Min Value}} \)
   - If the maximum and minimum values are equal (i.e., no variance), the value is set to 1 (or 100% for a KPI).

### 13. **Weighted Composite Score**
   - A **composite score** is calculated by combining the normalized 'Delta Median' and 'Reschedule Ratio', weighted by `W1` and `W2`:
     - \( \text{Composite} = \frac{W1 \times \text{Norm Delta} + W2 \times \text{Norm Reschedule}}{W1 + W2} \)
   - This composite score is a weighted average that combines both metrics into one value.

### 14. **Normalize Composite Score to 0-100 KPI**
   - The composite score is then normalized to a range of 0-100, so that it can be interpreted as a percentage-based KPI score.
   - If all scores are identical, the KPI is set to 100.

### 15. **Ensure KPI is within Bounds**
   - The KPI values are clipped to ensure they remain between 0 and 100, and rounded to two decimal places.

### 16. **Plotting the Results**
   - A plot is generated to visualize the KPI scores over time (month-by-month) for each team (`equipaggio`).
   - The plot shows the performance of each team in terms of their efficiency, as represented by the KPI scores.
   - The plot includes markers for each month's score, a title, axis labels, and a legend.

### 17. **Final Results Output**
   - The resulting DataFrame `kpi_df` is printed, showing the month, team, KPI score, Delta Median, and Reschedule Ratio for each team in each month.

This process allows you to compute a composite KPI for each team, considering both the median delta of assembly time and the ratio of rescheduled deliveries due to various issues. The final result provides a visual and numerical representation of each teamâ€™s efficiency over time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import iqr

# Rename column to remove new line
df.columns = [col.replace("\n", " ") for col in df.columns]

# Initialize results dictionary
monthly_kpis = []

# Iterate over each month and each equipaggio
for (month, equipaggio), df_group in df.groupby(['Mese', 'Id Equipaggio']):
    
    # Ensure there are valid delta values
    delta_values = df_group['Delta (orario di fine montaggio EFFETTIVO - orario di fine montaggio CONCORDATO) [min]'].dropna()
    
    if delta_values.empty:
        continue  # Skip if there are no valid values
    
    # Step 1: Handle Outliers in Delta
    q1, q3 = np.percentile(delta_values, [25, 75])
    iqr_value = iqr(delta_values)
    lower_bound = q1 - 1.5 * iqr_value
    upper_bound = q3 + 1.5 * iqr_value

    # Winsorize: Cap extreme values
    filtered_delta = delta_values.clip(lower=lower_bound, upper=upper_bound)

    # Step 2: Calculate Absolute Median Delta
    delta_median = abs(filtered_delta.median())

    # Step 3: Calculate Rescheduling Impact
    df_group['Total Reschedules'] = (
        df_group['# consegne ripianificate per problemi nel MONTAGGIO di "TIPO 1" (camere, armadio e camerette)'] +
        df_group['# consegne ripianificate per problemi nel MONTAGGIO di "TIPO 2" (cucine e soggiorni)'] +
        df_group['# consegne ripianificate per problemi nel MONTAGGIO  di "TIPO 3" (arredo bagno, tavoli, sedie)']
    )
    
    df_group['Total Reschedules_'] = (
        df_group['# consegne ripianificate per problemi nel TRASPORTO'] +
        df_group['# consegne ripianificate per problemi nel MONTAGGIO di "TIPO 1" (camere, armadio e camerette)'] +
        df_group['# consegne ripianificate per problemi nel MONTAGGIO di "TIPO 2" (cucine e soggiorni)'] +
        df_group['# consegne ripianificate per problemi nel MONTAGGIO  di "TIPO 3" (arredo bagno, tavoli, sedie)'] +
        df_group['# consegne ripianificate per CAUSE LATO CLIENTE FINALE (assenza) e AZIENDA COMMITTENTE (es. pezzi mancanti)']
    )
    
    

    
    total_served = df_group['# clienti  serviti nel mese'].sum()
    total_reschedules = df_group['Total Reschedules'].sum()
    total_reschedules_ = df_group['Total Reschedules_'].sum()
    reschedule_ratio = total_reschedules / (total_reschedules_ + total_served) if (total_reschedules + total_served) > 0 else 0

    # Store base metrics for later normalization
    monthly_kpis.append({
        'Mese': month,
        'Id Equipaggio': equipaggio,
        'Delta Median': delta_median,
        'Reschedule Ratio': reschedule_ratio
    })

# Convert to DataFrame for normalization
kpi_df = pd.DataFrame(monthly_kpis).sort_values(by=['Mese', 'Id Equipaggio'])

# Calculate normalization parameters
delta_min = kpi_df['Delta Median'].min()
delta_max = kpi_df['Delta Median'].max()
resched_min = kpi_df['Reschedule Ratio'].min()
resched_max = kpi_df['Reschedule Ratio'].max()

# Define weights (adjust based on business priorities)
W1 = 10  # Weight for Delta Median
W2 = 20  # Weight for Reschedule Ratio

# Normalize metrics and calculate composite score
kpi_df['Norm Delta'] = kpi_df['Delta Median'].apply(
    lambda x: (delta_max - x) / (delta_max - delta_min) if delta_max != delta_min else 1.0
)
kpi_df['Norm Reschedule'] = kpi_df['Reschedule Ratio'].apply(
    lambda x: (resched_max - x) / (resched_max - resched_min) if resched_max != resched_min else 1.0
)

# Calculate weighted composite score
kpi_df['Composite'] = (W1 * kpi_df['Norm Delta'] + W2 * kpi_df['Norm Reschedule']) / (W1 + W2)

# Normalize composite score to 0-100 KPI
composite_min = kpi_df['Composite'].min()
composite_max = kpi_df['Composite'].max()

if composite_max != composite_min:
    kpi_df['KPI'] = 100 * (kpi_df['Composite'] - composite_min) / (composite_max - composite_min)
else:
    kpi_df['KPI'] = 100  # All scores are identical

# Ensure KPI is within bounds
kpi_df['KPI'] = kpi_df['KPI'].clip(0, 100).round(2)

# Step 5: Plot Monthly KPI Values per Equipaggio
plt.figure(figsize=(12, 6))
for equipaggio in kpi_df['Id Equipaggio'].unique()[:5]:
    subset = kpi_df[kpi_df['Id Equipaggio'] == equipaggio]
    plt.plot(subset['Mese'], subset['KPI'], marker='o', linestyle='-', label=f'Equipaggio {equipaggio}')

plt.ylim(0, 100)
plt.xlabel("Month")
plt.ylabel("Efficiency Score")
plt.title("Assembly Efficiency KPI by Month and Equipaggio")
plt.legend()
plt.grid()
plt.show()

# Print results
print(kpi_df[['Mese', 'Id Equipaggio', 'KPI', 'Delta Median', 'Reschedule Ratio']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Clean missing data (e.g., drop rows where necessary)
df['# clienti  serviti nel mese'] = df['# clienti  serviti nel mese'].fillna(0)
df['# giorni lavorati nel mese'] = df['# giorni lavorati nel mese'].fillna(0)

# Convert relevant columns to numeric
df['# clienti  serviti nel mese'] = df['# clienti  serviti nel mese'].astype(int)
df['# giorni lavorati nel mese'] = df['# giorni lavorati nel mese'].astype(int)

# Calculate the rescheduled deliveries for all types
df['total_rescheduled'] = (
    df['# consegne ripianificate per problemi nel TRASPORTO'] +
    df['# consegne ripianificate per problemi nel MONTAGGIO di "TIPO 1" (camere, armadio e camerette)'] +
    df['# consegne ripianificate per problemi nel MONTAGGIO di "TIPO 2" (cucine e soggiorni)'] +
    df['# consegne ripianificate per problemi nel MONTAGGIO  di "TIPO 3" (arredo bagno, tavoli, sedie)'] +
    df['# consegne ripianificate per CAUSE LATO CLIENTE FINALE (assenza) e AZIENDA COMMITTENTE (es. pezzi mancanti)']
)

# Group by month (Mese) and calculate KPIs
monthly_data = df.groupby('Mese').agg(
    richieste_consegna=('total_rescheduled', 'sum'),  # Calculate total rescheduled deliveries for each month
    richieste_servite=('# clienti  serviti nel mese', 'sum'),
    risorse_umane=('# giorni lavorati nel mese', 'sum'),
    risorse_mezzi=('Id Equipaggio', 'nunique')  # count of unique teams (transport resources)
).reset_index()

# Calculate 'Richieste di Consegna' as requests served minus rescheduled requests
monthly_data['richieste_consegna'] = monthly_data['richieste_servite'] + monthly_data['richieste_consegna']

# Plotting
plt.figure(figsize=(14, 8))

# Plot each KPI
plt.subplot(2, 2, 1)
plt.plot(monthly_data['Mese'], monthly_data['richieste_consegna'], marker='o', color='b', label='Richieste Consegna')
plt.title('Richieste di Consegna per Mese')
plt.xlabel('Mese')
plt.ylabel('Numero di Richieste')

plt.subplot(2, 2, 2)
plt.plot(monthly_data['Mese'], monthly_data['richieste_servite'], marker='o', color='g', label='Richieste Servite')
plt.title('Numero di Richieste Servite per Mese')
plt.xlabel('Mese')
plt.ylabel('Numero di Richieste Servite')

plt.subplot(2, 2, 3)
plt.plot(monthly_data['Mese'], monthly_data['risorse_umane'], marker='o', color='r', label='Risorse Umane')
plt.title('Risorse Umane (Giorni Lavorati) per Mese')
plt.xlabel('Mese')
plt.ylabel('Numero di Giorni Lavorati')

plt.subplot(2, 2, 4)
plt.plot(monthly_data['Mese'], monthly_data['risorse_mezzi'], marker='o', color='purple', label='Risorse Mezzi')
plt.title('Risorse Mezzi (Equipaggi) per Mese')
plt.xlabel('Mese')
plt.ylabel('Numero di Mezzi')

plt.tight_layout()
plt.show()
```
